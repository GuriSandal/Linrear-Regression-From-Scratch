```python
import numpy as np

# Linear Regression with One variable

X = np.array([1,2,3,4,5,6])
y = np.array([2,3,4,5,6,7])

theta_0=0
theta_1=0
alpha=0.01


for i in range(len(X)):
    prediction = theta_0 + theta_1 * X[i]

    cost = (1/(2*len(X)))*(prediction - y[i])**2
    
    for j in range(28):
        temp_0 = theta_0 - alpha*(1/len(X))*(prediction - y[i])
        temp_1 = theta_1 - alpha*(1/len(X))*(prediction - y[i])*X[i]
        
        theta_0 = temp_0
        theta_1 = temp_1
        
        print("theta_0 = ",theta_0)
        print("theta_1 = ",theta_1)
        
        prediction = theta_0 + theta_1 * X[i]
    
        cost = (1/(2*len(X)))*(prediction - y[i])**2
        print("cost_loss = ",cost)
        print("\n")

        
```

```python
prediction = [theta_0 + theta_1 * i for i in X]
print(prediction)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.plot(X,prediction)
plt.show()
```
